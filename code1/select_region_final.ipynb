### 모아어린이집을 시행할 지역 채택하기 위해 데이터 만들기 & 최종 지역 선택하기

#### 데이터 만들기

```python
import pandas as pd
import re
```

```python
df1 = pd.read_csv('일반 현황_20221.csv',encoding='euc-kr',error_bad_lines=False)
df2 = pd.read_csv('행정구역_시군구_별_1세별_주민등록인구_20220802102945.csv',encoding='euc-kr',
                 names=['행정구역(시군구)별','연령별','총인구수 (명)','남자인구수 (명)','여자인구수 (명)'])
df3 = pd.read_csv('시도별_신고접수_건수_20220803095522.csv',encoding='euc-kr',
                 names=['시도별','계','응급아동학대 의심사례','아동학대 의심사례'])
```

```python
### 어린이집 국공립/사립 구분을 위한 '행정구역(시군구)별','설립유형' 을 이용하여 그룹화
df1 = df1[['유치원명','주소','설립유형']]
df1['설립유형'] = df1['설립유형'].str[0:2]
df1.loc[df1['설립유형'] == '공립', '설립유형'] = '국립'  # 공립, 국립을 '국립' 하나로 보기 위함
df1['행정구역(시군구)별'] = df1['주소'].str.split(" ").str[0]
drop_index = df1[df1['행정구역(시군구)별'] == '시흥시'].index  # 주소가 이상한 데이터 제거
df1 = df1.drop(drop_index)
df1 = df1.drop(['주소'],axis=1)
df1 = df1.groupby(['행정구역(시군구)별','설립유형']).count().reset_index()
df1
```

```python
df1_1 = df1.pivot(index='행정구역(시군구)별', columns='설립유형', values='유치원명').reset_index()
df1_1['어린이집'] = df1_1['국립'] + df1_1['사립']
df1_1 
```

```python
# 지역별 어린이집 수
'''
df1 = df1.sort_values('지역별').reset_index()
df1 = df1[['지역별','2021']]
df1.rename(columns = {'지역별' : '지역별','2021':'어린이집'}, inplace = True)

drop_index = df1[df1['지역별'] == '전국'].index # 주소가 이상한 데이터 제거
df1 = df1.drop(drop_index)'''
```

```python
# df2 전처리
df2 = df2.iloc[2:]
df2['연령별'] = df2['연령별'].str.extract(r'(\d+)')
df2 = df2.dropna()
drop_index = df2[df2['행정구역(시군구)별'] == '전국'].index  # 주소가 이상한 데이터 제거
df2 = df2.drop(drop_index)

cols = ['연령별','총인구수 (명)','남자인구수 (명)','여자인구수 (명)']
for col in cols:
    df2[col] = df2[col].astype('int')

df2
```

```python
# 시군구별 아동인구
df2_1 = df2[(df2['연령별']>=0)&(df2['연령별']<=5)]
df2_1.rename(columns = {'총인구수 (명)' : '아동인구(명)'}, inplace = True)
df2_1 = df2_1['아동인구(명)'].groupby(df2_1['행정구역(시군구)별']).sum().reset_index()
df2_1
```

```python
# df3 전처리
df3 =df3.iloc[2:]
df3= df3.dropna()
drop_index = df3[df3['시도별'] == '총계'].index  # 주소가 이상한 데이터 제거
df3 = df3.drop(drop_index)

df3 = df3[['시도별','계']]
df3 = df3.sort_values('시도별').reset_index()
df3.rename(columns = {'계' : '아동학대'}, inplace = True)
df3['아동학대'] = df3['아동학대'].astype('int')

df3
```

```python
df_selection = pd.concat([df1_1,df2_1,df3],axis=1)
df_selection.columns
```

```python
df_selection = pd.concat([df1_1,df2_1,df3],axis=1)
df_selection = df_selection[['행정구역(시군구)별','어린이집','아동인구(명)','국립', '사립','아동학대']]  # '행정구역(시군구)별' 컬럼이 두개 들어가는 것 확인
df_selection.columns = ['행정구역(시군구)별', '행정구역(시군구)별2','어린이집','아동인구(명)','국립', '사립','아동학대']  # 행정구역 컬럼명을 다르게 지정
df_selection = df_selection.drop(['행정구역(시군구)별2'],axis=1)  # 행정구역 컬럼 한개 삭제
df_selection
```

```python
# 2021 사회조사보고서(복지2-1) - 향후 늘려야 할 공공시설로 국공립어린이집을 늘려야 한다는 비율 데이터 넣기
daycare_center = [9.9, 13.7, 11.6, 8.2, 10.1, 10.7, 11.5, 9.6, 14, 11.7, 9.6, 12.1, 8.9, 9.8, 10, 9.3, 9.1]
df_selection['향후 늘려야 할 공공시설(국공립어린이집)'] = daycare_center
```

```python
# 2021 사회조사보고서(복지3-1) - 향후 늘려야 할 복지서비스로 보육서비스를 늘려야 한다는 비율 데이터 넣기
child_care_service = [3.9, 3.9, 3.5, 3.7, 3.7, 3.8, 3.6, 2.9, 4.6, 5.4, 4, 2.5, 3.3, 3.9, 4.7, 3.6, 4]
df_selection['향후 늘려야 할 복지 서비스(보육 관련 서비스)'] = child_care_service
```

```python
df_selection
```

```python
# 파생변수 만들기
df_selection['국립/사립'] = df_selection['국립'] / df_selection['사립']
df_selection['아동학대건수/아동인구수'] = df_selection['아동학대'] / df_selection['아동인구(명)']
df_selection['아동인구수/어린이집수'] = df_selection['아동인구(명)'] / df_selection['어린이집']
df_selection = df_selection.round({'국립/사립':2, '아동학대건수/아동인구수':4, '아동인구수/어린이집수':2})  # 여러 컬럼에 반올림 함수 적용
df_selection
```

#### 최종 지역 선택하기

```python
# rank 함수를 이용해 6개의 기준 각각에 대한 시도별 순위 구하기
# method='min' 사용: 동점 관측치 그룹 내 최소 순위 부여
# rank 함수는 기본적으로 오름차순 정렬. ascending=False를 입력하면 내림차순 정렬

df_selection['(기준1)국립/사립'] = df_selection['국립/사립'].rank(method='min')  # 오름차순 정렬
df_selection['(기준2)향후 국공립 필요성'] = df_selection['향후 늘려야 할 공공시설(국공립어린이집)'].rank(method='min', ascending=False)
df_selection['(기준3)향후 보육 서비스 필요성'] = df_selection['향후 늘려야 할 복지 서비스(보육 관련 서비스)'].rank(method='min', ascending=False)
df_selection['(기준4)아동학대건수/아동인구수'] = df_selection['아동학대건수/아동인구수'].rank(method='min', ascending=False)
df_selection['(기준5)아동인구수'] = df_selection['아동인구(명)'].rank(method='min', ascending=False)
df_selection['(기준6)아동인구수/어린이집수'] = df_selection['아동인구수/어린이집수'].rank(method='min', ascending=False)
```

```python
df_selection
```

```python
# 6개의 기준 각각에서의 순위를 합한 값으로 구성된 종합점수 컬럼 추가

df_selection['종합점수'] = df_selection.iloc[:, 11:].sum(axis=1)

# 종합점수 컬럼에 대해 rank 함수 적용
# 순위값 합이므로 값이 낮을수록 지역의 중요성이 큰 것 -> 오름차순 정렬(default)

df_selection['종합순위'] = df_selection['종합점수'].rank(method='min')
```

```python
df_selection = df_selection.sort_values('종합순위')
df_selection
```
